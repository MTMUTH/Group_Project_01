```python
dfqb = df.loc[df['position']=='QB']
dfrb = df.loc[df['position']=='RB']
dfwr = df.loc[df['position']=='WR']


dfqb = dfqb.sort_values(by=['positionrank'])
dfrb = dfrb.sort_values(by=['positionrank'])
dfwr = dfwr.sort_values(by=['positionrank'])

dftop =df[0:5]
dftop =df[5: ]

print(dfqb.columns)
```
