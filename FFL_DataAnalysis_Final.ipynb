```python
#WR    
for i in range(1,17):
    conn = http.client.HTTPSConnection('api.fantasydata.net')
    conn.request("GET", "/v3/nfl/stats/json/GameLeagueLeaders/2017/"+str(i)+"/WR/FantasyPoints?%s" % params, "{body}", headers)
    response = conn.getresponse()
    dataTWR= json.loads(response.read().decode('utf-8'))
    
    points = np.zeros_like(dfwr['position'])
    
    for j in dataTWR:
        name = j['Name']
        index= np.where(dfwr['player'] == name)
        #print(len(index[0]))
        if len(index[0]) > 0:
            points[index[0][0]] = j['FantasyPoints']
            
    dfwr['Wk'+str(i)] = points
```

```python
#RB    
for i in range(1,17):
    conn = http.client.HTTPSConnection('api.fantasydata.net')
    conn.request("GET", "/v3/nfl/stats/json/GameLeagueLeaders/2017/"+str(i)+"/RB/FantasyPoints?%s" % params, "{body}", headers)
    response = conn.getresponse()
    dataTRB = json.loads(response.read().decode('utf-8'))
    
    points = np.zeros_like(dfrb['position'])
    
    for j in dataTRB:
        name = j['Name']
        index= np.where(dfrb['player'] == name)
        #print(len(index[0]))
        if len(index[0]) > 0:
            points[index[0][0]] = j['FantasyPoints']
            
    dfrb['Wk'+str(i)] = points
```

# Season Averages and Delta Graphs

```python
#Remove Inactive Players in non-bye weeks
dfqb = dfqb.loc[(dfqb['Wk1'] > 0.0) & (dfqb['Wk4'] > 0.0) & (dfqb['Wk12'] > 0.0) & (dfqb['Wk14'] > 0.0)]
dfrb = dfrb.loc[(dfrb['Wk1'] > 0.0) & (dfrb['Wk4'] > 0.0) & (dfrb['Wk12'] > 0.0) & (dfrb['Wk14'] > 0.0)]
dfwr = dfwr.loc[(dfwr['Wk1'] > 0.0) & (dfwr['Wk4'] > 0.0) & (dfwr['Wk12'] > 0.0) & (dfwr['Wk14'] > 0.0)]
```

```python
#Season Averages by Position by creating data frames for top 5 and field by each position
dfqb5 = dfqb.iloc[:5] 
dfqbF = dfqb.iloc[5:]

dfwr5 = dfwr.iloc[:5]
dfwrF = dfwr.iloc[5:]

dfrb5 = dfrb.iloc[:5]
dfrbF = dfrb.iloc[5:]

T5QBSAvg = []
TFQBSAvg = []
T5WRSAvg = []
TFWRSAvg = []
T5RBSAvg = []
TFRBSAvg = []

for i in range(1,17):
    T5QBSAvg.append(np.mean(dfqb5['Wk'+str(i)].loc[dfqb5['Wk' + str(i)] > 0.0]))
    TFQBSAvg.append(np.mean(dfqbF['Wk'+str(i)].loc[dfqbF['Wk' + str(i)] > 0.0]))
    T5WRSAvg.append(np.mean(dfwr5['Wk'+str(i)].loc[dfwr5['Wk' + str(i)] > 0.0]))
    TFWRSAvg.append(np.mean(dfwrF['Wk'+str(i)].loc[dfwrF['Wk' + str(i)] > 0.0]))
    T5RBSAvg.append(np.mean(dfrb5['Wk'+str(i)].loc[dfrb5['Wk' + str(i)] > 0.0]))
    TFRBSAvg.append(np.mean(dfrbF['Wk'+str(i)].loc[dfrbF['Wk' + str(i)] > 0.0]))

#Test QB Season Averages
print(T5QBSAvg)
```

```python
#Creating Data Frames for Simulated Draft graphs
savgqb = []  #creating an array
savgrb = []
savgwr = []

swavgqb = []  #creating an array for weekly average by position
swavgrb = []
swavgwr = []

for i in range(10): # setting the loop based on how many samples desired by person running the test
#     sdqb = dfqb.sample(n = 1)  #data frame
#     sdwr = dfwr.sample(n = 2)
#     sdrb = dfrb.sample(n = 2)

    sdqb = dfqbF.sample(n = 1)  #data frame
    sdwr = dfwrF.sample(n = 2)
    sdrb = dfrbF.sample(n = 2)
    
    for i in range(1,17):  #sum the weekly sample totals by position
        swavgqb.append(np.sum(sdqb['Wk'+str(i)].loc[sdqb['Wk' + str(i)] > 0.0])) 
        swavgwr.append(np.sum(sdwr['Wk'+str(i)].loc[sdwr['Wk' + str(i)] > 0.0]))
        swavgrb.append(np.sum(sdrb['Wk'+str(i)].loc[sdrb['Wk' + str(i)] > 0.0]))
    
    savgqb.append(np.mean(swavgqb))
    savgwr.append(np.mean(swavgwr))
    savgrb.append(np.mean(swavgrb))
    
# print(savgqb)
# print(savgwr)
# print(savgrb)
```

```python
#Simulated Draft for the Remainder of the Field Graph
fig6 = plt.figure(figsize = (20,6))
sampledraft = fig6.add_subplot(111)
#print(T5QBSAvg,TFQBSAvg)
#print(np.subtract(np.array(T5QBSAvg), np.array(TFQBSAvg)))
sampledraft.plot(savgqb, color = 'red', label = 'Quarterback')
sampledraft.plot(savgwr, color = 'blue', label = 'Wide Receivers')
sampledraft.plot(savgrb, color = 'green', label = 'Running Backs')


sampledraft.set_xlabel('Number of Samples', size = 18)
sampledraft.set_ylabel('Avg Points', size = 18)
sampledraft.set_title('Draft Model for Fantasy Football - Remainder', size = 18)
sampledraft.legend(loc = 'best')


plt.savefig('fig6.png')
```
